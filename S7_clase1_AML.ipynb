# Coeficiente de asimetría de Fisher

El coeficiente de asimetría de Fisher es un coeficiente que permite calcular la asimetría de una distribución.  
Es decir, el coeficiente de asimetría sirve para determinar si una función es asimétrica positiva, asimétrica negativa o simétrica.  

Dada una muestra $$\{x_1, x_2, \ldots, x_n\}, $$ la fórmula del coeficiente de asimetría de Fisher es:

$$
\gamma_1 = \frac{\sum_{i=1}^{n}(x_i - \bar{x})^3}{n s^3}
$$

Donde $ \bar{x} $ y  s  son la media y la desviación estándar muestrales.



*   $
\gamma_1 >0 $ la distribución es asimétrica positiva. (cola a la derecha)
*    $
\gamma_1 < 0 $ la distribución es asimétrica negativa. (cola a la izquierda)

# Curtosis (Kurtosis)

La **curtosis (Kurtosis)** describe el grado en que las colas de una distribución difieren de las colas de una distribución normal.  
En otras palabras, mide qué tan **afilada o achatada** es la distribución en comparación con la normal.





Python (a través de *pandas*, *numpy* y *scipy*) calcula la **curtosis de Fisher**, también conocida como **exceso de curtosis**, mediante la siguiente fórmula:

$$
K = \frac{n(n+1)}{(n-1)(n-2)(n-3)}
\sum_{i=1}^{n}\left( \frac{x_i - \bar{x}}{s} \right)^4
- \frac{3(n-1)^2}{(n-2)(n-3)}
$$

Donde:
- $ K $ es la curtosis (exceso de curtosis),
- $ \bar{x} $ es la media muestral,
- $ s $ es la desviación estándar muestral,
- $ n $ es el tamaño de la muestra.




## Interpretación del coeficiente de curtosis en Python

| Tipo de curtosis | Valor de \( K \) (exceso de curtosis) | Características |
|------------------|----------------------------------------|-----------------|
| **Leptocúrtica** | \( K > 0 \) | Colas más pesadas y un pico más alto y estrecho que la normal. |
| **Mesocúrtica**  | \( K = 0 \) | Similar a la distribución normal (curtosis estándar). |
| **Platicúrtica** | \( K < 0 \) | Colas más ligeras y un pico más achatado que la normal. |

En esta definición, la **distribución normal** tiene \( K = 0 \).  



## Ejemplo gráfico ilustrativo

Forma general de las distribuciones según su curtosis:

- **Leptocúrtica** → colas más gruesas (valores extremos más probables).  
- **Mesocúrtica** → comportamiento normal.  
- **Platicúrtica** → colas más ligeras (valores extremos menos probables).

$$
\text{Leptocúrtica } > \text{ Mesocúrtica } > \text{ Platicúrtica}
$$

# Coeficiente de correlación de Pearson

El **coeficiente de correlación de Pearson (r)** mide la **fuerza y dirección de la relación lineal** entre dos variables cuantitativas.  
Su valor siempre se encuentra en el rango -1 a 1.

---

## Fórmula teórica

$$
r = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}
{\sqrt{\sum_{i=1}^{n} (x_i - \bar{x})^2} \sqrt{\sum_{i=1}^{n} (y_i - \bar{y})^2}}
$$

donde:

- $x_i, y_i$ son los valores de las variables \(X\) y \(Y\),
- $\bar{x}, \bar{y}$ son las medias de \(X\) y \(Y\),
- $n$ es el número de observaciones.

---

## Interpretación de los valores de \( r \)

| Valor aproximado de \( r \) | Tipo de correlación | Descripción |
|------------------------------|---------------------|--------------|
| $ r \approx 1 $ | **Positiva fuerte** | A medida que una variable aumenta, la otra también tiende a aumentar. |
| $ 0.5 \leq r < 1 $ | **Positiva moderada** | Existe una tendencia positiva, aunque no perfecta. |
|$ -0.5 < r < 0.5 $ | **Débil o nula** | No hay una relación lineal evidente entre las variables. |
| $ -1 < r \leq -0.5 $ | **Negativa moderada** | A medida que una variable aumenta, la otra tiende a disminuir. |
| $ r \approx -1 $ | **Negativa fuerte** | Relación lineal inversa casi perfecta. |

---

## Supuestos del Coeficiente de Correlación de Pearson


Para que el resultado del coeficiente de Pearson sea confiable, se deben cumplir ciertas condiciones o *supuestos estadísticos*.  
Estos aseguran que el valor de \( r \) refleje correctamente la relación entre las variables.


### 1. Nivel de medición
Las variables deben ser **numéricas y continuas** (por ejemplo: temperatura, presión, concentración, altura, peso).  
Esto significa que deben poder medirse en una escala con valores reales, no solo categorías o rangos.


### 2. Datos pareados
Cada observación debe tener un **par de valores** $(x_i, y_i)$.  
Es decir, los datos de las dos variables deben pertenecer al **mismo caso o experimento**.  
Si falta un dato en alguna de las dos variables, esa observación no debe usarse en el cálculo.


### 3. Normalidad bivariada
Los datos de ambas variables deben seguir una **distribución normal conjunta**.


### 4. Relación lineal
El coeficiente de Pearson **solo mide relaciones lineales**.  
Esto significa que los puntos deben poder aproximarse por una **línea recta**.  
Si la relación tiene forma de curva (por ejemplo, parabólica), Pearson no es el método adecuado.


### 5. Homocedasticidad
La **dispersión de los puntos** debe ser **similar en todo el rango de valores**.  
La varianza (dispersión de los datos) es constante en todas las observaciones de un conjunto de datos.
Si la dispersión cambia mucho, la correlación puede ser engañosa.



---

# Prueba de Normalidad de Shapiro–Wilk

La **prueba de Shapiro–Wilk** se utiliza para evaluar si un conjunto de datos proviene de una distribución normal.  
Es una de las pruebas más utilizadas y confiables para comprobar la normalidad, especialmente cuando se trabaja con **muestras pequeñas o moderadas** (n < 2000).

---

## Hipótesis del test

Esta prueba parte de una hipótesis nula que afirma que la muestra **sí proviene** de una distribución normal.

- **Hipótesis nula (H₀):**  
  La muestra **proviene de una distribución normal**.  
  Se **acepta H₀** si el valor *p* es **mayor que 0.05**.

- **Hipótesis alternativa (H₁):**  
  La muestra **no proviene de una distribución normal**.  
  Se **rechaza H₀** (y se acepta H₁) si el valor *p* es **menor que 0.05**.

---

## Interpretación del valor *p*

| Valor *p* | Interpretación |
|------------|----------------|
| \( p > 0.05 \) | No hay evidencia suficiente para rechazar la normalidad. Los datos se pueden considerar **normales**. |
| \( p < 0.05 \) | Hay evidencia suficiente para rechazar la normalidad. Los datos **no son normales**. |

Un valor *p* menor a 0.05 indica que los datos **no siguen una distribución normal**, y por lo tanto, **no se cumplen los supuestos del coeficiente de Pearson**.


````python
from scipy.stats import shapiro
````

````python
columnas = ['ozono', 'dioxido_azufre', 'dioxido_nitrogeno', 'monoxido_carbono', 'PM10']

for col in columnas:
    estadistico, p_value = shapiro(contam[col])
    print(f'{col}:  Estadístico = {estadistico:.3f}, p-value = {p_value:.10f}\n')
````

Si el p-valor es mayor que 0.05, podemos asumir que los datos de ese contaminante son normales.

Si el p-valor es menor que 0.05, los datos no son normales, y se recomienda usar una correlación no paramétrica (como Spearman o Kendall) en lugar de Pearson.

# Correlación de Spearman $\rho$

# Correlación no paramétrica de Spearman

Cuando los datos **no cumplen el supuesto de normalidad**, no es adecuado usar la **correlación de Pearson**, ya que esta se basa en distribuciones normales y relaciones lineales.  
En estos casos, se recomienda aplicar la **correlación de Spearman**, también conocida como **Rho de Spearman (ρ)**.

---

La **correlación de Spearman** evalúa la **fuerza y dirección de la relación monótona** (creciente o decreciente) entre dos variables numéricas u ordinales.  
A diferencia de Pearson, **no requiere que los datos sean normales ni que la relación sea lineal**.

Spearman convierte los valores de las variables en **rangos (posiciones ordenadas)** y calcula la correlación entre esos rangos.

---

## Fórmula teórica

$$
\rho = 1 - \frac{6 \sum_{i=1}^{n} d_i^2}{n(n^2 - 1)}
$$

donde:

- $ d_i $ es la diferencia entre los rangos de las observaciones de \( X \) y \( Y \),
-  n  es el número de pares de observaciones.

El resultado, $\rho$, siempre se encuentra entre -1 y 1.

---

## Interpretación de los valores de ρ

| Valor aproximado de ρ | Tipo de correlación | Descripción |
|------------------------|---------------------|--------------|
| $ \rho \approx 1 $ | **Positiva fuerte** | A medida que una variable aumenta, la otra también lo hace. |
| $ 0.5 \leq \rho < 1 $ | **Positiva moderada** | Aumento de una variable tiende a acompañarse del aumento de la otra. |
| $ -0.5 < \rho < 0.5 $ | **Débil o nula** | No hay una relación monótona clara. |
| $ -1 < \rho \leq -0.5 $ | **Negativa moderada** | A medida que una variable aumenta, la otra tiende a disminuir. |
| $ \rho \approx -1 $ | **Negativa fuerte** | Relación inversa casi perfecta. |

---

## Supuestos del coeficiente de Spearman

Aunque Spearman es más flexible que Pearson, también tiene algunas condiciones a considerar:

1. **Tipo de datos:**  
   Las variables deben ser **numéricas u ordinales** (se pueden ordenar).  

2. **Relación monótona:**  
   No es necesario que la relación sea lineal, pero sí debe ser **monótona**:  
   cuando una variable aumenta, la otra **debe aumentar o disminuir consistentemente**.

3. **Datos pareados:**  
   Cada observación debe tener un par de valores $(x_i, y_i)$, como en Pearson.

4. **Ausencia de valores atípicos extremos:**  
   Aunque Spearman es menos sensible que Pearson, valores muy atípicos aún pueden afectar el resultado.

---

# Correlación vs Causalidad

En estadística, es muy común confundir **correlación** con **causalidad**,  
pero son conceptos **diferentes**.  

---

##  ¿Qué es la correlación?

La **correlación** mide **qué tan relacionadas están dos variables**:  
es decir, si cambian juntas y en qué dirección lo hacen.

- Si una variable **aumenta cuando la otra aumenta**, la correlación es **positiva**.  
- Si una **aumenta mientras la otra disminuye**, la correlación es **negativa**.  
- Si **no hay relación**, la correlación es **nula**.

 *Ejemplo:*  
Cuando aumenta la concentración de **ozono (O₃)**, también aumenta la de **PM10**.  
→ Existe una **correlación positiva**, pero eso no significa que una cause a la otra.

---

##  ¿Qué es la causalidad?

La **causalidad** implica que **un cambio en una variable provoca un cambio en otra**.  
Es decir, existe una **relación de causa y efecto**.

 *Ejemplo:*  
Encender una estufa **causa** un aumento en la temperatura.  
Aquí sí hay causalidad, porque una acción genera directamente el cambio.

---

##  Diferencias clave

| Aspecto | Correlación | Causalidad |
|----------|--------------|-------------|
| Qué mide | Asociación entre variables | Efecto directo de una variable sobre otra |
| Dirección | No implica dirección (A se asocia con B) | Tiene dirección (A causa B) |
| Comprobación | Se calcula con datos (Pearson, Spearman, etc.) | Se demuestra con experimentos o modelos causales |
| Ejemplo | Temperatura y consumo de helado aumentan juntos  | Encender una estufa **causa** aumento de temperatura  |




---

##  Correlación **no implica** causalidad

Que dos variables estén relacionadas **no significa que una cause a la otra**.  
A veces ambas dependen de una **tercera variable oculta** (llamada variable de confusión).

*Ejemplo clásico:*  
- Cuando hace calor, aumenta el **consumo de helado**   
- También aumentan los **casos de insolación**  
→ Helado e insolación están correlacionados,  
pero **la causa real** es la **temperatura**.

---

##  Caso de los contaminantes

Supón que al analizar tus datos encuentras que **ozono (O₃)** y **dióxido de nitrógeno (NO₂)**  
tienen una **alta correlación positiva**.

Eso **no significa** que el ozono cause el aumento de NO₂.  
Ambos pueden aumentar al mismo tiempo porque:

- Se generan bajo **condiciones similares** (como tráfico o radiación solar).  
- Provienen de **fuentes de emisión comunes** (vehículos, combustión, etc.).

 Por tanto:
> Hay correlación entre O₃ y NO₂,  
> pero no necesariamente una relación de causa y efecto.

---

##

| Concepto | Qué indica | Cómo se comprueba |
|-----------|-------------|-------------------|
| **Correlación** | Dos variables cambian juntas | Cálculos estadísticos (Pearson, Spearman, etc.) |
| **Causalidad** | Una variable provoca el cambio en otra | Experimentos o modelos causales |
| **Relación entre ambas** | Toda causalidad implica correlación, pero **no toda correlación implica causalidad** | Depende del contexto y del análisis |

---
 **Conclusión:**
- La **correlación** nos dice si hay una relación entre variables.  
- La **causalidad** nos dice si una realmente produce el cambio en la otra.  
- En tus datos de **contaminantes**, la correlación te muestra qué variables se mueven juntas,  
  pero para afirmar causalidad necesitarías un análisis más profundo o un experimento controlado.

# Pearson vs Spearman

````python

from scipy.stats import spearmanr, pearsonr

# Datos de ejemplo (no lineales, pero monótonos)
x = [1, 2, 3, 4, 5, 6, 7, 8, 9]
y = [2, 4, 6, 8, 10, 12, 14, 16, 30]  # crece en general, pero con un valor extremo al final

# Crear un DataFrame
df = pd.DataFrame({'X': x, 'Y': y})

# Calcular las correlaciones
r_pearson, p_pearson = pearsonr(df['X'], df['Y'])
r_spearman, p_spearman = spearmanr(df['X'], df['Y'])

print(f"Coeficiente de Pearson:  r = {r_pearson:.3f}, p = {p_pearson:.4f}")
print(f"Coeficiente de Spearman:  ρ = {r_spearman:.3f}, p = {p_spearman:.4f}")
````

El valor de Pearson (0.956) no llega a 1 porque hay un valor extremo (30) que rompe la linealidad.

En cambio, Spearman (1.000) detecta que, aunque los datos no sean perfectamente lineales, mantienen un orden creciente perfecto → por eso da 1.0

````python
plt.scatter(df['X'], df['Y'], color='royalblue')
plt.title('Relación entre X y Y (monótona pero no lineal)')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(alpha=0.3)
plt.show()
````
