# Coeficiente de asimetría de Fisher

El coeficiente de asimetría de Fisher es un coeficiente que permite calcular la asimetría de una distribución.  
Es decir, el coeficiente de asimetría sirve para determinar si una función es asimétrica positiva, asimétrica negativa o simétrica.  

Dada una muestra $$\{x_1, x_2, \ldots, x_n\}, $$ la fórmula del coeficiente de asimetría de Fisher es:

$$
\gamma_1 = \frac{\sum_{i=1}^{n}(x_i - \bar{x})^3}{n s^3}
$$

Donde $ \bar{x} $ y  s  son la media y la desviación estándar muestrales.



*   $
\gamma_1 >0 $ la distribución es asimétrica positiva. (cola a la derecha)
*    $
\gamma_1 < 0 $ la distribución es asimétrica negativa. (cola a la izquierda)

# Curtosis (Kurtosis)

La **curtosis (Kurtosis)** describe el grado en que las colas de una distribución difieren de las colas de una distribución normal.  
En otras palabras, mide qué tan **afilada o achatada** es la distribución en comparación con la normal.





Python (a través de *pandas*, *numpy* y *scipy*) calcula la **curtosis de Fisher**, también conocida como **exceso de curtosis**, mediante la siguiente fórmula:

$$
K = \frac{n(n+1)}{(n-1)(n-2)(n-3)}
\sum_{i=1}^{n}\left( \frac{x_i - \bar{x}}{s} \right)^4
- \frac{3(n-1)^2}{(n-2)(n-3)}
$$

Donde:
- $ K $ es la curtosis (exceso de curtosis),
- $ \bar{x} $ es la media muestral,
- $ s $ es la desviación estándar muestral,
- $ n $ es el tamaño de la muestra.




## Interpretación del coeficiente de curtosis en Python

| Tipo de curtosis | Valor de \( K \) (exceso de curtosis) | Características |
|------------------|----------------------------------------|-----------------|
| **Leptocúrtica** | \( K > 0 \) | Colas más pesadas y un pico más alto y estrecho que la normal. |
| **Mesocúrtica**  | \( K = 0 \) | Similar a la distribución normal (curtosis estándar). |
| **Platicúrtica** | \( K < 0 \) | Colas más ligeras y un pico más achatado que la normal. |

En esta definición, la **distribución normal** tiene \( K = 0 \).  



## Ejemplo gráfico ilustrativo

Forma general de las distribuciones según su curtosis:

- **Leptocúrtica** → colas más gruesas (valores extremos más probables).  
- **Mesocúrtica** → comportamiento normal.  
- **Platicúrtica** → colas más ligeras (valores extremos menos probables).

$$
\text{Leptocúrtica } > \text{ Mesocúrtica } > \text{ Platicúrtica}
$$

````python

# Calcular la curtosis de Fisher (exceso de curtosis) para cada contaminante
curtosis = contam.kurt()

# Mostrar los resultados
print("\nCurtosis de Fisher:\n")
print(curtosis)
````

dioxido_nitrogeno → tiene una distribución leptocúrtica (muchos valores extremos).

ozono y monoxido_carbono → platicúrticas (más planas, menos valores extremos).

PM10 → ligeramente leptocúrtica.

````python
# --- 1. Calcular curtosis de Fisher ---
curtosis = contam.kurt()
# --- 2. Crear los histogramas ---
columnas = contam.columns

plt.figure(figsize=(15, 8))  # tamaño del lienzo

for i, col in enumerate(columnas, 1):
    plt.subplot(2, 3, i)  # 2 filas, 3 columnas de gráficos
    plt.hist(contam[col], bins=20, color='skyblue', edgecolor='black')
    plt.title(f"{col}\nCurtosis: {curtosis[col]:.2f}", fontsize=12)
    plt.xlabel(col)
    plt.ylabel('Frecuencia')


plt.suptitle("Distribución de contaminantes y su curtosis (Fisher)", fontsize=16)
plt.tight_layout()
plt.show()
````

# Coeficiente de correlación de Pearson

El **coeficiente de correlación de Pearson (r)** mide la **fuerza y dirección de la relación lineal** entre dos variables cuantitativas.  
Su valor siempre se encuentra en el rango -1 a 1.

---

## Fórmula teórica

$$
r = \frac{\sum_{i=1}^{n} (x_i - \bar{x})(y_i - \bar{y})}
{\sqrt{\sum_{i=1}^{n} (x_i - \bar{x})^2} \sqrt{\sum_{i=1}^{n} (y_i - \bar{y})^2}}
$$

donde:

- $x_i, y_i$ son los valores de las variables \(X\) y \(Y\),
- $\bar{x}, \bar{y}$ son las medias de \(X\) y \(Y\),
- $n$ es el número de observaciones.

---

## Interpretación de los valores de \( r \)

| Valor aproximado de \( r \) | Tipo de correlación | Descripción |
|------------------------------|---------------------|--------------|
| $ r \approx 1 $ | **Positiva fuerte** | A medida que una variable aumenta, la otra también tiende a aumentar. |
| $ 0.5 \leq r < 1 $ | **Positiva moderada** | Existe una tendencia positiva, aunque no perfecta. |
|$ -0.5 < r < 0.5 $ | **Débil o nula** | No hay una relación lineal evidente entre las variables. |
| $ -1 < r \leq -0.5 $ | **Negativa moderada** | A medida que una variable aumenta, la otra tiende a disminuir. |
| $ r \approx -1 $ | **Negativa fuerte** | Relación lineal inversa casi perfecta. |

---

## Supuestos del Coeficiente de Correlación de Pearson


Para que el resultado del coeficiente de Pearson sea confiable, se deben cumplir ciertas condiciones o *supuestos estadísticos*.  
Estos aseguran que el valor de \( r \) refleje correctamente la relación entre las variables.


### 1. Nivel de medición
Las variables deben ser **numéricas y continuas** (por ejemplo: temperatura, presión, concentración, altura, peso).  
Esto significa que deben poder medirse en una escala con valores reales, no solo categorías o rangos.


### 2. Datos pareados
Cada observación debe tener un **par de valores** $(x_i, y_i)$.  
Es decir, los datos de las dos variables deben pertenecer al **mismo caso o experimento**.  
Si falta un dato en alguna de las dos variables, esa observación no debe usarse en el cálculo.


### 3. Normalidad bivariada
Los datos de ambas variables deben seguir una **distribución normal conjunta**.


### 4. Relación lineal
El coeficiente de Pearson **solo mide relaciones lineales**.  
Esto significa que los puntos deben poder aproximarse por una **línea recta**.  
Si la relación tiene forma de curva (por ejemplo, parabólica), Pearson no es el método adecuado.


### 5. Homocedasticidad
La **dispersión de los puntos** debe ser **similar en todo el rango de valores**.  
La varianza (dispersión de los datos) es constante en todas las observaciones de un conjunto de datos.
Si la dispersión cambia mucho, la correlación puede ser engañosa.



---

# Prueba de Normalidad de Shapiro–Wilk

La **prueba de Shapiro–Wilk** se utiliza para evaluar si un conjunto de datos proviene de una distribución normal.  
Es una de las pruebas más utilizadas y confiables para comprobar la normalidad, especialmente cuando se trabaja con **muestras pequeñas o moderadas** (n < 2000).

---

## Hipótesis del test

Esta prueba parte de una hipótesis nula que afirma que la muestra **sí proviene** de una distribución normal.

- **Hipótesis nula (H₀):**  
  La muestra **proviene de una distribución normal**.  
  Se **acepta H₀** si el valor *p* es **mayor que 0.05**.

- **Hipótesis alternativa (H₁):**  
  La muestra **no proviene de una distribución normal**.  
  Se **rechaza H₀** (y se acepta H₁) si el valor *p* es **menor que 0.05**.

---

## Interpretación del valor *p*

| Valor *p* | Interpretación |
|------------|----------------|
| \( p > 0.05 \) | No hay evidencia suficiente para rechazar la normalidad. Los datos se pueden considerar **normales**. |
| \( p < 0.05 \) | Hay evidencia suficiente para rechazar la normalidad. Los datos **no son normales**. |

Un valor *p* menor a 0.05 indica que los datos **no siguen una distribución normal**, y por lo tanto, **no se cumplen los supuestos del coeficiente de Pearson**.


````python
from scipy.stats import shapiro

for col in contam.columns:
    stat, p = shapiro(contam[col])
    print(f"{col}: Estadístico = {stat:.3f}, p-valor = {p:.8f}")
````

Si el p-valor es mayor que 0.05, podemos asumir que los datos de ese contaminante son normales.

Si el p-valor es menor que 0.05, los datos no son normales, y se recomienda usar una correlación no paramétrica (como Spearman o Kendall) en lugar de Pearson.
