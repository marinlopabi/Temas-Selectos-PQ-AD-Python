# Funciones




Una función es un conjunto de instrucciones que se agrupan para realizar una tarea específica.

- Recibe datos (entrada), los procesa y devuelve un resultado (salida).

*Sirven para reutilizar código, organizar mejor los programas y evitar repetir instrucciones.*


| Concepto       | Significado                             |
| -------------- | --------------------------------------- |
| **def**        | palabra clave para definir una función  |
| **Parámetros** | datos que la función recibe             |
| **return**     | indica qué valor devuelve la función    |
| **Llamada**    | cuando usamos la función en el programa |


```python
def sumar(a, b):  #La función recibe dos números, realiza una operación y devuelve el resultado.
    resultado = a + b
    return resultado
```

```python
sumar(4324,234) #Llamar la función suma
```

```python
def resta(a,b):
    res = a - b
    return res
```

 Código que acepte temperatura y presión en distintas unidades y determine el estado del agua.

| Condición           | Estado del agua          |
| ------------------- | ------------------------ |
| T < Tc y P < Pc     | Líquido o vapor          |
| T > Tc y P > Pc     | Supercrítico             |
| T < Tc y P > Pc     | Líquido comprimido       |
| Otras combinaciones | Cercano al punto crítico |


Paso 1: Funciones de conversión

Conversiones T

Función que cambie de C a K, F a K



```python

def a_K(T, unidad):
    if unidad == "C":
       res= T +273.15
       return res
    elif unidad == "F":
      res = (T - 32)* 5/9 +273.15
      return res
    elif unidad == "R":
       res = T* 5/9
       return res
    else:
      print("Dato no válido, verifique las unidades")
```

```python
a_K(25, "F")
```

```python
a_K(25, "h")
```
