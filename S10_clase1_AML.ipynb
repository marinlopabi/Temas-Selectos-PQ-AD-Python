```python
sector_map = {
    1: "CRUZ ROJA",
    2: "DIF",
    3: "ESTATAL",
    4: "IMSS",
    5: "IMSS-BIENESTAR",
    6: "ISSSTE",
    7: "MUNICIPAL",
    8: "PEMEX",
    9: "PRIVADA",
    10: "SEDENA",
    11: "SEMAR",
    12: "SSA",
    13: "UNIVERSITARIO",
    14: "CIJ",
    15: "IMSS Bienestar OPD",
    99: "NO ESPECIFICADO"
}
```

```python
df['SECTOR'] = df['SECTOR'].map(sector_map)
```

```python
df['SECTOR']
```

```python
df['SEXO'] = df['SEXO'].map({1: 'MUJER', 2: 'HOMBRE',3:'NO ESPECIFICADO'}) #Recodificar
```

```python
df['TIPO_PACIENTE'] = df['TIPO_PACIENTE'].map({1:'AMBULATORIO', 2:'HOSPITALIZADO',3:'NO ESPECIFICADO'})

```

```python
clasificacion_covid_map = {
    1: "CONFIRMADO",
    2: "CONFIRMADO",
    3: "CONFIRMADO",
    4: "NO VÁLIDO",
    5: "NO VÁLIDO",
    6: "SOSPECHOSO",
    7: "NEGATIVO"
}
```

```python
df['CLASIFICACION_FINAL_COVID'] = df['CLASIFICACION_FINAL_COVID'].map(clasificacion_covid_map)
```

```python
df['CLASIFICACION_FINAL_COVID']
```

```python
#Convertir Fechas a datos de tiempo
fechas = ['FECHA_SINTOMAS', 'FECHA_INGRESO','FECHA_DEF', 'FECHA_ACTUALIZACION']
for f in fechas:
    df[f] = pd.to_datetime(df[f], errors='coerce')
```

```python
df['FECHA_DEF']
```

```python
# Lista de columnas que se desean eliminar del DataFrame
# Columnas no son necesarias para el análisis posterior
columnas_eliminar = ['ORIGEN', 'ENTIDAD_RES', 'ID_REGISTRO', 'ENTIDAD_UM', 'FECHA_DEF', 'OTRO_CASO']

# Elimina las columnas especificadas del DataFrame
# axis=1: indica que se eliminarán columnas (no filas)
# errors='ignore': evita errores si alguna columna no existe en el DataFrame
df = df.drop(columnas_eliminar, axis=1, errors='ignore')
df
```

```python
#Conteo por categoría
df['SEXO'].value_counts()
```

```python
conteo_sexo = df['SEXO'].value_counts()
plt.figure(figsize=(6,4))

# Genera un gráfico de barras con los conteos de sexo
# kind='bar': tipo de gráfico de barras verticales
# color: asigna colores personalizados a cada barra (azul acero y salmón)
conteo_sexo.plot(kind='bar', color=['steelblue', 'salmon'])

# Configura el título principal
plt.title('Número de pacientes por sexo')

# Etiqueta del eje X (horizontal)
plt.xlabel('Sexo')

# Etiqueta del eje Y (vertical)
plt.ylabel('Cantidad de ingresos')

plt.xticks(rotation=0)

# Añade líneas de cuadrícula horizontales punteadas con transparencia
# axis='y': solo en el eje vertical
# linestyle='--': líneas discontinuas
# alpha=0.5: transparencia al 50%
plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
```

```python
plt.figure(figsize=(6,6))

# Genera un gráfico de pastel (pie chart) con los conteos de sexo
# kind='pie': tipo de gráfico circular
# autopct='%1.1f%%': muestra porcentajes con 1 decimal
# startangle=90: inicia el gráfico desde 90° (parte superior)
# colors: asigna colores personalizados a cada sección (azul acero y salmón)
conteo_sexo.plot(kind='pie', autopct='%1.1f%%', startangle=90, colors=['steelblue', 'salmon'])

# Elimina la etiqueta del eje Y (no tiene sentido en gráficos circulares)
plt.ylabel('')

plt.title('Proporción de ingresos por sexo')

plt.show()
```

```python
# Filtra el DataFrame para obtener solo los casos confirmados de COVID-19
# Crea un nuevo DataFrame con únicamente las filas donde CLASIFICACION_FINAL_COVID == 'CONFIRMADO'
df_positivos = df[df['CLASIFICACION_FINAL_COVID'] == 'CONFIRMADO']

# Agrupa los datos por fecha de ingreso y sexo, cuenta los casos y pivotea la tabla
# .groupby(['FECHA_INGRESO', 'SEXO']): agrupa por dos columnas
# .size(): cuenta el número de registros en cada grupo
# .unstack(): convierte SEXO de filas a columnas (pivotea), creando una columna por cada sexo
# .plot(): genera un gráfico de líneas con los datos pivoteados
df_positivos.groupby(['FECHA_INGRESO', 'SEXO']).size().unstack().plot(figsize=(10,5))


plt.title('Ingresos diarios por sexo')

plt.ylabel('Número de casos')

plt.show()
```

```python
# Filtra el DataFrame para obtener solo los casos confirmados
df_conf = df[df['CLASIFICACION_FINAL_COVID'] == 'CONFIRMADO']

plt.figure(figsize=(8,5))
plt.hist(df_conf['EDAD'], bins=30, color='steelblue', edgecolor='black', alpha=0.7)

plt.title('Distribución de edad en casos confirmados de COVID-19')

# Etiqueta del eje X
plt.xlabel('Edad')

# Etiqueta del eje Y
plt.ylabel('Número de casos')

# Añade líneas de cuadrícula horizontales punteadas con transparencia
plt.grid(axis='y', linestyle='--', alpha=0.4)

plt.show()
```

```python
# Cuenta la frecuencia de cada edad, ordena por edad (no por frecuencia) y grafica
# .value_counts(): cuenta cuántos casos hay para cada edad
# .sort_index(): ordena por edad (índice) de menor a mayor, no por cantidad de casos
# .plot(): genera el gráfico de barras
# kind='bar': tipo de gráfico de barras verticales
# color='darkcyan': color cian oscuro para las barras
df_conf['EDAD'].value_counts().sort_index().plot(
    kind='bar', figsize=(14,5), color='darkcyan'
)

plt.title('Casos confirmados por edad')
plt.xlabel('Edad')
plt.ylabel('Número de casos')

plt.show()
```

```python
# Define los límites de los intervalos para agrupar edades
# Crea 5 grupos: [0-17], [18-29], [30-49], [50-64], [65-120]
bins = [0, 17, 29, 49, 64, 120]

# Define las etiquetas legibles para cada grupo de edad
labels = ['0–17', '18–29', '30–49', '50–64', '65+']

# Crea una nueva columna 'grupo_edad' categorizando cada edad en su grupo correspondiente
# pd.cut(): función que divide datos numéricos en intervalos/bins
# bins=bins: usa los límites definidos arriba
# labels=labels: asigna las etiquetas a cada intervalo
# right=True: el límite derecho está incluido (ej: 17 pertenece a '0-17', no a '18-29')
df_conf['grupo_edad'] = pd.cut(df_conf['EDAD'], bins=bins, labels=labels, right=True)


df_conf['grupo_edad'].value_counts().sort_index().plot(
    kind='bar', color='orchid', figsize=(7,5)
)

plt.title('Casos confirmados por grupo de edad')

plt.xlabel('Grupo de edad')

plt.ylabel('Número de casos')


plt.show()
```
