Ahora añadimos más temperaturas

```python
def T_kelvin(T, unidad):
    if unidad == "C":
        return T + 273.15
    elif unidad == "F":
        return (T - 32)* 5/9 + 273.15
    elif unidad == "R":
        return T* 5/9
    elif unidad == "K":
        return T
    else:
       raise ValueError("Unidad de temperatura no válida. Usa C, F, R o K.")

```

```python
capitales = ['#Ciudad Victo*ria', 'cuerna#vaca', 'Morelia!', 'xalapa', 'Tepic', 'hermos*illo',
 'toluca?', 'San? Luis Potosí', '*Mexicali', 'Villahermosa', 'Tuxtla Gutiérrez#',
 'Colima', 'la Paz#', 'Guadalajara', 'C*ampeche!', 'Durango', 'Zacatecas',
 '!Chetumal', 'Ciudad# de México', 'chilpancingo*', 'Chihua#hua', '?Aguascalientes',
 'Oaxaca de Juárez*', '?Puebla', 'Quer?étaro', 'Mérida', 'guanajuato',
 'pachu*ca', '#Saltillo#', 'Culiacán!', 'Tlaxcala', 'monterrey']
```

Una expresión regular (o regex, por regular expression) es una secuencia de caracteres que define un patrón de búsqueda en cadenas de texto. Se usa para buscar, validar, reemplazar o dividir texto de forma eficiente y flexible.

Una forma de limpiar la lista es utilizando métodos de cadena de texto internos junto con el módulo de librería estándar **re** para las expresiones regulares:

```python
import re

def limpiar_cadenas(strings):
    result = []
    for value in strings:
        value = value.strip()  # suprime blancos u otro carácter especificado del final o del principio de una expresión
        value = re.sub('[#*?!]', '', value) #reemplaza todas las ocurrencias de un patrón por el texto que especifiquemos
        value = value.title() #convierte la primera letra de cada palabra de una cadena a mayúsculas.
        result.append(value)
    return result
```

```python
limpiar_cadenas(capitales)
```

```python
capitales_limpias = limpiar_cadenas(capitales)

capitales_limpias.sort() #sort para ordenar en orden Alfabético
```

# Sistema de ecuaciones

Resuelve el siguiente sistema de ecuaciones aplicando el método de Cramer:

$$
\begin{cases}
2x_1 + 4x_2 + 6x_3 = 18, \\
4x_1 + 5x_2 + 6x_3 = 24, \\
3x_1 + x_2 - 2x_3 = 4.
\end{cases}
$$

1. Identifica la matriz de coeficientes (A)

```python
A = np.array([[2, 4, 6],
              [4, 5, 6],
              [3, 1, -2]])
```

2. Calcula el determinante principal (D o det(A))

```python
D=np.linalg.det(A)
D
```

3. Crea matrices modificadas para cada variable

- Para encontrar **X**: reemplaza la primera columna de A con el vector de términos independientes.
- Para encontrar **Y**: reemplaza la segunda columna
- Para encontrar **Z**: reemplaza la tercera columna (y así sucesivamente)

```python
X = np.array([[18,4,6],
               [24, 5,6],
               [4, 1, -2]])
```

4. Calcula el determinante de cada matriz modificada

Dx, Dy, Dz, etc.

```python
Dx = np.linalg.det(X)
Dx
```

5. Aplica la fórmula de Cramer

- x = Dx/D

- y = Dy/D

- z = Dz/D

```python
x = Dx/D
x
```

```python
b = np.array([18,24,4])  #Comprobación

x=np.linalg.solve(A, b)
x
```

```python
np.dot(A,x)
```
