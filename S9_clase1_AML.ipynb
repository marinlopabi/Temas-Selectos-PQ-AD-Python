# **Archivos CSV**

| **Argumento**      | **Descripción**                                                                                   |
|--------------------|------------------------------------------------------------------------------------------------------------|
| `path`             | Ruta del archivo (local o URL). También puede ser un objeto tipo archivo o buffer.                        |
| `sep` / `delimiter`| Separador de campos. Por defecto `','`. Usa `\t` para TSV, o una expresión regular como `'\s+'` para espacios. |
| `header`           | Índice de fila a usar como nombres de columnas. Usa `header=None` si no hay encabezado.                   |
| `index_col`        | Columna(s) a usar como índice del DataFrame. Puede ser nombre, número o lista.                            |
| `names`            | Lista de nombres de columnas. Útil si no hay encabezado (`header=None`).                                   |
| `skiprows`         | Número de filas (o lista de índices) a omitir al inicio. Ideal para saltar metadatos o comentarios.        |
| `na_values`        | Lista de strings que serán interpretados como NA/NaN. Ej: `["n/a", "NA", "--"]`                            |
| `keep_default_na`  | Si `False`, desactiva los valores NA predeterminados (como `'NA'`, `'null'`, etc.).                       |
| `comment`          | Carácter que indica que el resto de la línea debe ignorarse como comentario. Ej: `comment='#'`            |
| `parse_dates`      | Intenta convertir columnas a datetime. Usa `True`, una lista de columnas o combinaciones de columnas.     |
| `keep_date_col`    | Si `True`, mantiene las columnas originales usadas para crear fechas.                                      |
| `converters`       | Diccionario de funciones para aplicar a columnas específicas. Ej: `{"col1": lambda x: x.strip()}`         |
| `dayfirst`         | Si `True`, interpreta fechas ambiguas como día/mes/año. (por ej. 10/11/2022 → 11 de octubre)              |
| `date_parser`      | Función personalizada para convertir fechas. Se usa con `parse_dates`.                                     |
| `nrows`            | Número de filas a leer desde el inicio (sin contar encabezado).                                            |
| `iterator`         | Devuelve un iterador (`TextFileReader`) para leer por fragmentos (útil con `chunksize`).                   |
| `chunksize`        | Tamaño de los bloques (en filas) que se leen por iteración. Ideal para archivos muy grandes.              |
| `skip_footer`      | Número de filas a ignorar desde el final del archivo. Requiere `engine='python'`.                         |
| `verbose`          | Muestra información adicional durante la carga del archivo (útil para depuración).                         |
| `encoding`         | Codificación del archivo. Por ejemplo: `"utf-8"`, `"latin1"`, `"cp1252"`                                   |
| `squeeze`          | Si el archivo tiene una sola columna, la devuelve como `Series` en vez de `DataFrame`.                    |
| `thousands`        | Separador de miles (como `','` o `'.'`) para convertir strings a números correctamente.                    |
| `decimal`          | Separador decimal, por defecto `"."`. Usa `","` para archivos europeos.                                   |
| `engine`           | Motor de lectura: `"c"` (rápido por defecto), `"python"` (más flexible), `"pyarrow"` (experimental).       |

**La codificación**

- es la forma en que las computadoras convierten texto en números para poder guardarlo y leerlo. Cada carácter (letra, número, símbolo) tiene un código numérico.

| Codificación     | Nombre Completo                             | Caracteres Soportados                                             | Uso Principal                                      | Ventajas                                                     | Desventajas                                                 |
|------------------|----------------------------------------------|-------------------------------------------------------------------|----------------------------------------------------|---------------------------------------------------------------|-------------------------------------------------------------|
| **UTF-8**        | Unicode Transformation Format 8-bit          | Todos los idiomas, emojis, símbolos especiales (+1 millón)        | Estándar mundial moderno, web, APIs                | Universal, compatible con ASCII, eficiente en espacio        | Archivos antiguos pueden no usarlo                          |
| **Latin-1 (ISO-8859-1)** | Latin Alphabet No. 1                | Idiomas europeos occidentales (256 caracteres)                    | Archivos antiguos de Europa Occidental             | Simple, compatible con sistemas viejos                       | No soporta idiomas no europeos, obsoleto                   |
| **Windows-1252 (CP1252)** | Windows Code Page 1252             | Similar a Latin-1 con caracteres extra                            | Archivos de Excel y Windows en español/inglés      | Usado por defecto en Windows antiguo                         | Problemas de compatibilidad entre sistemas                 |
| **ASCII**        | American Standard Code for Information Interchange | Solo inglés básico (128 caracteres)                          | Programación, sistemas antiguos                    | Universal, muy simple                                        | No tiene acentos ni ñ, muy limitado                        |
| **UTF-16**       | Unicode Transformation Format 16-bit         | Todos los idiomas como UTF-8                                      | Sistemas internos de Windows, Java                 | Bueno para idiomas asiáticos                                 | Ocupa más espacio, menos común en archivos                 |
| **CP850**        | Code Page 850                                | DOS Europa Occidental                                              | Sistemas DOS antiguos, terminales                  | Soporta caracteres de caja (│, ─, ┌)                         | Obsoleto, solo para compatibilidad                         |
| **ISO-8859-15**  | Latin-9                                      | Como Latin-1 pero con símbolo €                                   | Europa después del euro                             | Incluye €, Œ, œ                                              | Menos usado que UTF-8                                      |

# **Archivos TXT**

Un archivo txt se puede leer con read_csv() porque ambos son solo texto delimitado. Pandas no se fija en la extensión, sino en cómo están separados los datos. Si el txt tiene comas, punto y coma, tabulaciones o espacios, read_csv() sabe interpretarlo como tabla.

| Extensión | Significa                | Estructura real                                            |
| --------- | ------------------------ | ---------------------------------------------------------- |
| `.csv`    | *comma separated values* | texto separado por comas, punto y coma, tabulaciones, etc. |
| `.txt`    | texto genérico           | puede contener cualquier cosa, incluso datos tabulados     |


read_csv() lee CSVs:

- .txt

- .dat

- .tsv

- archivos sin extensión

- cualquier archivo de texto que tenga delimitadores

## Códigos de tipos en NumPy/pandas:

| **Código** | **Significado**           | **Tipo completo**                  | **Ejemplo**                     |
|------------|----------------------------|-------------------------------------|----------------------------------|
| **O**      | Object (texto/mixto)       | `object`                            | `'Ana'`, `'Hola'`               |
| **i**      | Integer (entero)           | `int64`, `int32`                    | `5`, `100`, `-2`                |
| **f**      | Float (decimal)            | `float64`, `float32`                | `3.14`, `25.5`                  |
| **b**      | Boolean                    | `bool`                              | `True`, `False`                 |
| **U**      | Unicode string             | `str`, `unicode`                    | `"Hola"`, `"Mundo"`             |
| **M**      | Datetime                   | `datetime64[ns]`                    | `2024-01-15`                    |
| **m**      | Timedelta                  | `timedelta64[ns]`                   | `2 days`, `5 hours`             |
| **c**      | Complex number             | `complex64`, `complex128`           | `3+2j`                          |
| **V**      | Raw data (void)            | `void`                              | Datos binarios sin interpretar  |
| **S**      | Byte string (ASCII)        | `bytes`, `string`                   | `b"hola"`                       |
| **category** | Categórico               | `category`                           | `'Bajo'`, `'Medio'`, `'Alto'`  , Género: Masculino, Femenino, Otro |

## Convertir datos de un tipo a otro

```python
# Cambiar tipo de dato
df["fecha"] = pd.to_datetime(df["fecha"]) # Convertir a fecha
df["edad"] = df["edad"].astype(int)
df['edad'] = df['edad'].astype(float)
df['nombre'] = df['nombre'].astype('category')

# Seleccionar solo columnas numéricas
df.select_dtypes(include=['int64', 'float64'])

# Seleccionar solo columnas de texto
df.select_dtypes(include=['object'])

# Contar tipos de datos
df.dtypes.value_counts()
```

## Renombrar columnas

```python
df.rename(columns={"Temp": "Temperatura"}, inplace=True) #Renombra la columna "Temp" a "Temperatura" y modifica el DataFrame
```

```python
# Renombrar varias columnas a la vez
df.rename(columns={
    "Temp": "Temperatura",
    "Hum": "Humedad",
    "Pres": "Presión"
}, inplace=True)

print(df.columns.tolist())
# ['Temperatura', 'Humedad', 'Presión']
```

```python
# Renombrar TODAS las columnas (sobrescribir lista completa)
df.columns = ['Temperatura', 'Humedad', 'Presión']

# Usando la función upper() (convertir a mayúsculas) a cada nombre de columna.
df.rename(columns=str.upper, inplace=True)
# Temp → TEMP, Hum → HUM

df.columns = df.columns.str.lower()  # Minúsculas

# Añadir prefijo o sufijo
df.add_prefix('col_')   # col_Temp, col_Hum
df.add_suffix('_data')  # Temp_data, Hum_data

# Reemplazar espacios o caracteres
df.columns = df.columns.str.replace(' ', '_') #De espacios a guiones
# 'Nombre Completo' → 'Nombre_Completo'
```

## Duplicados

```python
df.duplicated().sum() #Detectar duplicados
```

```python
df.drop_duplicates(inplace=True) #Eliminar duplicados
```

```python
# Verificar duplicados solo en columnas específicas
df.duplicated(subset=['nombre']).sum()

# Marcar TODAS las apariciones como duplicadas (no solo las segundas)
df.duplicated(keep=False).sum()

# Mantener la última aparición en vez de la primera
df.duplicated(keep='last').sum()

# Ver cuáles son las filas duplicadas
df[df.duplicated()]

# Contar filas únicas (opuesto)
len(df) - df.duplicated().sum()
```
